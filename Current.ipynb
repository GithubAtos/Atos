```python
df_Import_Export = df_import_export.rename(columns={'Perioden':'Timeframe',
                                                   'BrutoProductie_1':'Total_Production',
                                                   'InvoerTotaal_17':'Total_Import',
                                                   'Belgie_18':'Import_Belgium',
                                                   'Denemarken_19':'Import_Denmark',
                                                   'Duitsland_20':'Import_Germany',
                                                   'VerenigdKoninkrijk_21':'Import_UK',
                                                   'Noorwegen_22':'Import_Norway',
                                                   'UitvoerTotaal_23':'Total_Export',
                                                   'Belgie_24':'Export_Belgium',
                                                   'Denemarken_25':'Export_Denmark',
                                                   'Duitsland_26':'Export_Germany',
                                                   'VerenigdKoninkrijk_27':'Export_UK',
                                                   'Noorwegen_28':'Export_Norway',
                                                   'Distributieverliezen_29':'Energy_lost'})
df_Import_Export.head(100)
```

```python
df_Import_Export.isnull().sum()
```

```python
df_Import_Export.fillna(0, inplace=True)
df_Import_Export.head(15)
```

```python
df_Import_Export.isnull().sum()
```

```python
df_Import_Export_year= df_Import_Export[df_Import_Export['Timeframe'].str.contains('JJ', na = False)]
df_Import_Export_year.tail(10)
```

```python
df_Import_Export_month= df_Import_Export[df_Import_Export['Timeframe'].str.contains('MM', na = False)]
df_Import_Export_month.tail(10)
```

```python
df_Import_Export_month['Net_Import']= df_Import_Export_month['Total_Import']-df_Import_Export_month['Total_Export']
df_Import_Export_month.tail(10) 
df_Import_Export_year['Net_Import']= df_Import_Export_year['Total_Import']-df_Import_Export_year['Total_Export']
df_Import_Export_year.tail(10)
```

```python
plt.plot(df_Import_Export_year['Timeframe'], df_Import_Export_year['Net_Import'], color='red', marker='o')
plt.title('Net Import over time', fontsize=10)
plt.xlabel('Year', fontsize=10)
plt.ylabel('GWh', fontsize=10)
plt.grid(False)
plt.show()
```
