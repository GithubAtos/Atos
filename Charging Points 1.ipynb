```python
#let's and some widgets and also let's make it interactive 

#Widget
select_widget=ipywidgets.Select(
    options=['Open Street Map', 'Terrain', 'Toner', 'Watercolor', 'Positron', 'Dark Matter'],
    value='Open Street Map',
    description='Map Type:',
    disabled=False)

# widget function
def select(map_type):
    if map_type == 'Open Street Map':
        display(folium.Map(location=[52.1326, 5.2913], zoom_start=7, height=700))
    if map_type == 'Terrain':
        display(folium.Map(location=[52.1326, 5.2913], tiles='Stamen Terrain', zoom_start=7, height=400))
    if map_type == 'Toner':
        display(folium.Map(location=[52.1326, 5.2913], tiles='Stamen Toner', zoom_start=7, height=400))
    if map_type == 'Watercolor':
        display(folium.Map(location=[52.1326, 5.2913], tiles='Stamen Watercolor', zoom_start=7, height=400))
    if map_type == 'Positron':
        display(folium.Map(location=[52.1326, 5.2913], tiles='CartoDB Positron', zoom_start=7, height=400))
    if map_type == 'Dark Matter':
        display(folium.Map(location=[52.1326, 5.2913], tiles='CartoDB Dark_Matter', zoom_start=7, height=400))
        
# interaction between widgets and function    
ipywidgets.interact(select, map_type=select_widget)
```

```python
#Let's add layer control to the map 
map_layer_control = folium.Map(location= [52, 5], zoom_start=7)

#add tiles to the map 

folium.raster_layers.TileLayer('Open Street Map').add_to(map_layer_control)
folium.raster_layers.TileLayer('Stamen Terrain').add_to(map_layer_control)
folium.raster_layers.TileLayer('Stamen Toner').add_to(map_layer_control)
folium.raster_layers.TileLayer('Stamen Watercolor').add_to(map_layer_control)
folium.raster_layers.TileLayer('CartoDB Positron').add_to(map_layer_control)
folium.raster_layers.TileLayer('CartoDB Dark_Matter').add_to(map_layer_control)

folium.LayerControl().add_to(map_layer_control)

map_layer_control
```

```python
#Mini map fore easy location through the map
map_with_mini = folium.Map(location =[52, 5], zoom_start=7) 
minimap= plugins.MiniMap(toggle_display=True) 

#add minimap to map 
map_with_mini.add_child(minimap)

#Add scroll zoom feature the map 
plugins.ScrollZoomToggler().add_to(map_with_mini)

#Now let's add fullscreen button to the screen 
plugins.Fullscreen(position='topright').add_to(map_with_mini)

map_with_mini

```

```python
#Let's try to plot the charging points on the map 

#create map 
ch_ports = folium.Map(location=[52, 5], zoom_start=7)

#plot locations
for (index, row) in df.iterrows():
    folium.Marker(location=[row.loc['Location(WKT)']]).add_to(ch_ports)

ch_ports
```
